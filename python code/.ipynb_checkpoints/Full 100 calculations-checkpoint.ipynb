### Trim TC image to only include bead

```python
def beadcreator(arr, low_threshold, high_threshold):
    masked_arr = np.ma.masked_less_equal(arr, low_threshold)
    
    thickness = np.ma.count(masked_arr, axis = 0)
    cropped_columns = np.where(thickness < 9)
    
    m = np.zeros_like(masked_arr)
    m[:, cropped_columns] = 1
    
    cropped_masked_arr = np.ma.masked_array(masked_arr, m)
    cropped_masked_arr = np.ma.masked_greater_equal(cropped_masked_arr, high_threshold)
    return cropped_masked_arr

fig, axes = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(9,8)

data = TC_1[1]
I1 = axes[0,0].imshow(data, interpolation='None', cmap = plt.cm.gnuplot, vmin = 0, vmax = 65535)
Green_TC_1 = beadcreator(data, 15000, 60000)
I2 = axes[0,1].imshow(Green_TC_1, interpolation='None', cmap=plt.cm.binary, vmin = 0, vmax = 65535)

fig.colorbar(I1, ax = axes[0,0])
fig.colorbar(I2, ax = axes[0,1])

data = TC_2[1]
I1 = axes[1,0].imshow(data, interpolation='None', cmap = plt.cm.gnuplot, vmin = 0, vmax = 65535)
Green_TC_2 = beadcreator(data, 15000, 62000)
I2 = axes[1,1].imshow(Green_TC_2, interpolation='None', cmap=plt.cm.binary, vmin = 0, vmax = 65535)

fig.colorbar(I1, ax = axes[1,0])
fig.colorbar(I2, ax = axes[1,1])

for x,y in itertools.product([0,1], [0,1]):
    axes[x,y].set_xticks([])
    axes[x,y].set_yticks([])

fig.tight_layout()
```

### Soot calculation function as per Cignotti et al.

```python
def e_TC(l):
    answer = 1.2018e-6 * (l*1e9)**2 - (l*1e9) * 1.7167e-3 + 0.9017
    return answer

def soot_fv_calculator(T_soot, green_soot, tau_soot):
    
    """ This function calculates the soot volume fraction
    using flame luminosity, soot temperature, and an 
    independent absolute light calibration. 
    
    INPUTS:
    -------
    T_soot: Numpy array : 2D map of soot temperature
    green_soot: Numpy array : BASEX inverted green channel of the flame image
    tau_soot: float : Exposure time of the flame image
    
    PARAMETERS:
    -----------
    S_TC_n : float : Luminosity of green channel of light calibration source
                     (default source is a Type-S thermocouple)
                     (n = 2 usually, i.e., images of the TC at 2 different temperatures is used)
    T_TC_n : float : Temperature of the light calibration source
                     (default is the self-reported TC temperature)
    tau_TC_n : float : Exposure time of the light calibration source image
    spa_rez : float : Spatial resolution of the camera setup in units of pixels per mm
                      (For e.g., spa_rez = 28 means 28 pixels for every mm)
    
    RETURNS:
    --------
    f_v_avg : Numpy array : 2D map of soot volume fraction 
                            at all physical locations for which both 
                            T_soot and green_soot information is available
    
                            Multiply the output by 1e6 to convert to ppm
    
    """
    
    lambda_S = 533e-9 # m, when used with e_TC(), it is converted to nm inside the function
    K_ext = 8.6
    L = 1e-3/spa_rez # m, dimension of 1 pixel, there being "spa_rez" pixels in every mm
    e_TC_at_lambda_S = e_TC(lambda_S)
    h = constants.Planck
    c = constants.c
    k = constants.Boltzmann
    
    f_v_1 = (-1.0*lambda_S/(K_ext*L))*np.log(1 - e_TC_at_lambda_S
                                * (tau_TC_1 / tau_soot)
                                * (green_soot / S_TC_1)
                                * np.exp((-1*h*c/(k*lambda_S))
                                            *(1/T_TC_1 - 1/T_soot)))

    f_v_2 = (-1.0*lambda_S/(K_ext*L))*np.log(1 - e_TC_at_lambda_S
                                * (tau_TC_2 / tau_soot)
                                * (green_soot / S_TC_2)
                                * np.exp((-1*h*c/(k*lambda_S))
                                            *(1/T_TC_2 - 1/T_soot)))
    
    f_v_avg = 0.5*(f_v_1 + f_v_2)
    
    return f_v_avg
```

### Process one flame

```python
t = time.time()
for index in [50]: #typically pick 50th image, because why not?
    all_flames = [red_flame_names[index], 
                  green_flame_names[index], 
                  blue_flame_names[index]]
    
    raw_flames = [sc.genfromtxt(x) for x in all_flames]
    
    smooth_flames = [csc.sgolay2d(x, 7,1) for x in raw_flames]
    
    inverted_flames = [inv_ab(x, 185, verbose=False) for x in smooth_flames]

    for x in inverted_flames:
        x[np.where(x < 0)] = 0
    
    center = inverted_flames[0].shape[1]//2

    for idx, x in enumerate(inverted_flames):
#         inverted_flames[idx][:,center-13:center+14] = ndimage.median_filter(x,
#                                                                             size = (27,27))[:,center-13:center+14]
#         inverted_flames[idx][:,center-13:center+14] = signal.medfilt2d(x,kernel_size = (27,27))[:,center-13:center+14]
    

        inverted_flames[idx][:,center-3:center+4] = ndimage.filters.median_filter(x, 
                                                                                  size = (1,9))[:,center-3:center+4]
        
    red_flame = inverted_flames[0] 
    green_flame = inverted_flames[1]
    blue_flame = inverted_flames[2]
    
    x = ndimage.filters.gaussian_filter(red_flame, 10, truncate=6.0)
    mask_loose = np.ma.masked_where(x < 12, np.ones_like(x))
#     mask_loose = np.ma.masked_where(x < 0.02*x.max(), np.ones_like(x))
#     mask_tight = np.ma.masked_where(x < 0.15*x.max(), np.ones_like(x))
    mask_tight = np.ma.masked_where(x < 12, np.ones_like(x))
    
    masked_red_flame = np.ma.masked_where(np.ma.getmask(mask_loose), red_flame)
    masked_green_flame = np.ma.masked_where(np.ma.getmask(mask_loose), green_flame)
    masked_blue_flame = np.ma.masked_where(np.ma.getmask(mask_loose), blue_flame)
    
    green_red = masked_green_flame / masked_red_flame
    blue_red = masked_blue_flame / masked_red_flame
    blue_green = masked_blue_flame / masked_green_flame
    
    red_temp = Interpolated_Lookup_bg(blue_green)
    green_temp = Interpolated_Lookup_br(blue_red)
    blue_temp = Interpolated_Lookup_gr(green_red)

    average_temp = (blue_temp+red_temp+green_temp)/3.0
    
    soot_T = np.ma.masked_where(np.ma.getmask(mask_tight), average_temp)
    green_soot = np.ma.array(inverted_flames[1], mask = soot_T.mask)
#     tau_soot = 1.0/2.0
    
    soot_fv = soot_fv_calculator(soot_T, green_soot, tau_soot)
    
    nan_soot_T = soot_T.filled(fill_value=np.nan)

    nan_soot_fv = soot_fv.filled(fill_value=np.nan)
    
#     np.savez_compressed('./data/flame_%d' %index, nan_soot_T=nan_soot_T, nan_soot_fv=nan_soot_fv)
elapsed = time.time() - t
print elapsed
```

### Process all flames

```python
centers = np.zeros((100,3))
center_location = []
problem_locations = []

for index in range(100):  #Note:range here was 100 
        
    all_flames = [red_flame_names[index], 
                  green_flame_names[index], 
                  blue_flame_names[index]]
    
    raw_flames = [sc.genfromtxt(x) for x in all_flames]
    
    centers[index] = np.array([x[np.where(x > np.percentile(x, 99.5))].mean() for x in raw_flames])
    center_location += [np.where(x == x.max()) for x in raw_flames]
    
    if raw_flames[1].max() >= 65535:
        problem_locations += [index] 

    smooth_flames = [csc.sgolay2d(x, 7,1) for x in raw_flames]
    
    inverted_flames = [inv_ab(x, 185,verbose=False) for x in smooth_flames] # Input center column from above
    
    for x in inverted_flames:
        x[np.where(x < 0)] = 0
    
    center = inverted_flames[0].shape[1]//2
    for idx, x in enumerate(inverted_flames):
        inverted_flames[idx][:,center-3:center+4] = ndimage.filters.median_filter(x, 
                                                                                  size = (1,9))[:,center-3:center+4]
        
    red_flame = inverted_flames[0] 
    green_flame = inverted_flames[1]
    blue_flame = inverted_flames[2]
    
    x = ndimage.filters.gaussian_filter(red_flame, 10, truncate=6.0)
    mask_loose = np.ma.masked_where(x < 12, np.ones_like(x))
#     mask_loose = np.ma.masked_where(x < 0.02*x.max(), np.ones_like(x))
#     mask_tight = np.ma.masked_where(x < 0.15*x.max(), np.ones_like(x))
    mask_tight = np.ma.masked_where(x < 12, np.ones_like(x))
    
    masked_red_flame = np.ma.masked_where(np.ma.getmask(mask_loose), red_flame)
    masked_green_flame = np.ma.masked_where(np.ma.getmask(mask_loose), green_flame)
    masked_blue_flame = np.ma.masked_where(np.ma.getmask(mask_loose), blue_flame)
    
    green_red = masked_green_flame / masked_red_flame
    blue_red = masked_blue_flame / masked_red_flame
    blue_green = masked_blue_flame / masked_green_flame
    
    red_temp = Interpolated_Lookup_bg(blue_green)
    green_temp = Interpolated_Lookup_br(blue_red)
    blue_temp = Interpolated_Lookup_gr(green_red)

    average_temp = (blue_temp+red_temp+green_temp)/3.0
    
    soot_T = np.ma.masked_where(np.ma.getmask(mask_tight), average_temp)
    green_soot = np.ma.array(inverted_flames[1], mask = soot_T.mask)
    
    soot_fv = soot_fv_calculator(soot_T, green_soot, tau_soot)
    
    nan_soot_T = soot_T.filled(fill_value=np.nan)
    nan_soot_fv = soot_fv.filled(fill_value=np.nan)
    
    np.savez_compressed(save_folder+'./saves/flame_%d' %index, 
                        nan_soot_T=nan_soot_T, nan_soot_fv=nan_soot_fv, center_v = centers[index])
    
    if index % 5 == 0:
        sys.stdout.write('{}...'.format(index))
        sys.stdout.flush()
        
print "\nDone!"
print "Problem locations:", problem_locations
```
